# **Import**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
plt.style.use('dark_background')
```

# **Penguins**

```python
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
penguins.head(1)
```

```python
df_penguins = sns.load_dataset('penguins')
df_penguins.head(1)
```

**Описание данных `df_penguins`**:

- **`species:`** вид пингвина (например, Adelie, Chinstrap, Gentoo)
- **`island:`** остров, на котором был найден пингвин (например, Torgersen, Dream, Biscoe)
- **`bill_length_mm:`** длина клюва в миллиметрах
- **`bill_depth_mm:`** глубина клюва в миллиметрах
- **`flipper_length_mm:`** длина плавника в миллиметрах
- **`body_mass_g:`** масса тела в граммах
- **`sex:`** пол пингвина (мужской, женский или неизвестный)

# **Diamonds**

```python
df_diamonds = sns.load_dataset('diamonds')
df_diamonds.head(1)
```

**Описание данных `df_diamonds`**:

- **`carat:`** вес бриллианта в каратах (1 карат = 0.2 грамма)
- **`cut:`** качество огранки (Fair, Good, Very Good, Premium, Ideal)
- **`color:`** цвет (D — бесцветный, далее E, F, G и т.д. по убыванию ценности)
- **`clarity:`** чистота (I1, SI2, SI1, VS2, VS1, VVS2, VVS1, IF)
- **`depth:`** глубина (процентное соотношение глубины к диаметру)
- **`table:`** ширина верхней грани (процент от общей ширины)
- **`price:`** цена в долларах США
- **`x:`** длина в миллиметрах
- **`y:`** ширина в миллиметрах
- **`z:`** глубина в миллиметрах
